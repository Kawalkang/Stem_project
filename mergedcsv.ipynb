```python
import statsmodels.api as sm
import pandas as pd

# Select the independent variables
X = df1[['Turn Count', 'Total Touches (#)', 'Top Speed (m/s)', 'Sprint Count (#)', 'Release Velocity Avg (m/s)', 'Work Rate (m/min)']]

# Calculate missing values
missing_values = X.isna().sum()

# Replace missing values with the mean of each column
X = X.fillna(X.mean())

# Response variable
Y = df1['Total Possessions (#)']

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(Y, X).fit()

# Print the summary of the regression model
print(model.summary())

# Extract the summary as text
summary_text = model.summary().as_text()

# Split the text by newline characters to separate rows
summary_lines = summary_text.split('\n')

# Initialize an empty DataFrame to store the summary information
summary_data = []

# Iterate through the lines, and split each line into cells
for line in summary_lines:
    cells = line.split()
    summary_data.append(cells)

# Create a Pandas DataFrame from the summary data
df_summary = pd.DataFrame(summary_data)

# Save the DataFrame to an Excel file
df_summary.to_excel(r'C:\Users\ASUS\Desktop\D. STEM PROJECT\df9.xlsx', index=False)
```
